```python
import matplotlib.pyplot as plt
import numpy as np

# LIMONENE:
limonene_solubility_trials_75 = {
    "Ethylene glycol": [0, 0, 0], 
    "Methanol": [40, 40, 40],
    "Acetic acid": [100, 80, 80],
    "Ethanol": [140, 140, 140],
    "DMA": [40, 20, 20]
}

limonene_solubility_trials_90 = {
    "Ethylene glycol": [0, 0, 0],
    "Methanol": [240, 240, 240],
    "Acetic acid": [340, 360, 360],
    "Ethanol": [1040, 1040, 1040],
    "DMA": [460, 480, 480]
}

solvent_ranks = {
    "Ethylene glycol": 17,
    "Methanol": 16,
    "Acetic acid": 9,
    "Ethanol": 6,
    "DMA": 2
}

solvents = list(limonene_solubility_trials_75.keys())
solubility_means_75 = [np.mean(limonene_solubility_trials_75[solvent]) for solvent in solvents]
solubility_means_90 = [np.mean(limonene_solubility_trials_90[solvent]) for solvent in solvents]

# Plotting
fig, ax = plt.subplots(figsize=(8, 6))
y = np.arange(len(solvents))
width = 0.35

rects1 = ax.barh(y - width/2, solubility_means_75, width, 
                 capsize=5, label='75%', 
                 color=['#6bff86', '#ff5c5c', '#74B9FF', '#fa9c50', '#CBA6F1'][::-1], 
                 edgecolor='black', linewidth=1.5)

rects2 = ax.barh(y + width/2, solubility_means_90, width, 
                 capsize=5, label='90%', 
                 color=['#02b823', '#e80505', '#00008B', '#d46002', '#a000a6'][::-1], 
                 edgecolor='black', linewidth=1.5)

ax.set_xlabel('Limonene concentration (mM)', fontsize=16, labelpad=18)
ax.set_yticks(y)
ranks = [solvent_ranks[solvent] for solvent in solvents]
ax.set_yticklabels([f'{solvent}\nRank: {rank}' for solvent, rank in zip(solvents, ranks)], fontsize=14)
ax.set_xlim(0, 1200)

def autolabel(rects, solubilities, percentage_label):
    for rect, solubility in zip(rects, solubilities):
        width = rect.get_width()
        if solubility == 0:
            label = f'insoluble {percentage_label}'
        else:
            label = f'{percentage_label}'
        ax.annotate(label,
                    xy=(width, rect.get_y() + rect.get_height() / 2),
                    xytext=(2, 0),  
                    textcoords="offset points",
                    ha='left', va='center', fontsize=14)

autolabel(rects1, solubility_means_75, '75V%')
autolabel(rects2, solubility_means_90, '90V%')

ax.tick_params(axis='y', rotation=45)
ax.tick_params(axis='x', labelsize=14)
ax.set_xlim(0, 1320)

plt.tight_layout()
plt.show()
```
